# Weather API
This file shows the ability to use an API from the Weather website and get data from it.

- Used the following libraries:
- To make an API request - requests
- To decode JSON - json

```python
# Import package to make HTTP requests, i.e. fetch URL similar to a browser request
import requests
# Import json library to JSON
import json
```

```python
api_url = 'https://api.weatherapi.com/v1/current.json'

params = {'key' : '******', 'q':'90405'}

# make API request
api_request = requests.get(api_url, params=params)
```
